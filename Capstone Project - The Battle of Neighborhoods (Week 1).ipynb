```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
# Getting data from wiki

url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
response = requests.get(url)

# Reading it using Beautifulsoup

data = response.text
soup = BeautifulSoup(data,'html.parser')
table = soup.find('table')
```

```python
table_contents = []

for row in table.findAll('td'):
    cell = {}
    if row.span.text=='Not assigned':
        pass
    else:
        cell['Postalcode'] = row.p.text[:3]
        cell['Borough'] = (row.span.text).split('(')[0]
        cell['Neighborhood'] = (((((row.span.text).split('(')[1]).strip(')')).replace(' /',',')).replace(')',' ')).strip(' ')
        table_contents.append(cell)
        
table_contents
```

```python
for i in table_contents:
    l=[]
    l = [i['Neighborhood']]
    if len(l)>1:
        for j in range(len(l)):
            i['Postalcode']=i['Postalcode']
            i['Borough']=i['Borough']
            i['Neighborhood']=l[j]
            
```

```python
for i in table_contents:
    if ',' in i['Neighborhood']:    
        l = i['Neighborhood'].split(',')   
        for j in range(len(l)):
            k={}
            k['Postalcode']=i['Postalcode']
            k['Borough']=i['Borough']
            k['Neighborhood']=l[j]
            table_contents.append(k)
        table_contents.remove(i) 
    
```
